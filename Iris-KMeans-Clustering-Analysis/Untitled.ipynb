```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#importing dataset
dataset=pd.read_csv("Irıs_Data_Set.csv")
```

```python
#statistics attributes
dataset.describe()
```

```python
#species
dataset["species"].unique()
```

```python
dataset.plot(kind="scatter", x="sepal_length",   y="sepal_width",color="red")
plt.show()
```

```python
dataset.plot(kind="scatter", x="petal_length",   y="petal_width",color="green")
plt.show()
```

```python

#Assigned value of the X for kmeans clustering-X[sepal_length-sepal_width]
#Assigned value of the Y for kmeans clustering-Y[petal_length-petal_width]
#Assigned value of the X for kmeans clustering-Z[All elements]
X=dataset.iloc[:,[0,1]].values
Z=dataset.iloc[:,[2,3]].values
A=dataset.iloc[:,[0,1,2,3]].values
```

```python
#Option 1 -2
```

```python
#************************SEPAL (Length-Width)**********************
```

```python
# finding the optimum number of clusters for - kmeans algorithm


wcss=[] # with in cluster sum of square error
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=200,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)#method
```

```python
#elbow method-plotting the result into a line graph
plt.plot(range(1,11),wcss)
plt.title('THE ELBOW METHOD')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#applying k-means algorithm to iris data set
#************************ K=3 **********************
kmeans=KMeans(n_clusters=3,init='k-means++',max_iter=200,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(X)
    
```

```python
# visualsing the clusters with scatter plot
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Setosa')

plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='green',label='Versicolor')

plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='blue',label='Virginica')

# plotting centroid of clusters

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='Centroids')

plt.title("K-means")
plt.legend()
```
